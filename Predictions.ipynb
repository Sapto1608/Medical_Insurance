```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
df=pd.read_csv(r"C:\Users\User\OneDrive\Desktop\machine_learning\Medicalpremium.csv")
```

```python
x= df.drop(columns='PremiumPrice',axis=1)
y=df['PremiumPrice']
```

```python
from sklearn.preprocessing import StandardScaler
scalar =  StandardScaler()
x.Age = scalar.fit_transform(x[['Age']])
x.Height = scalar.fit_transform(x[['Height']])
x.Weight = scalar.fit_transform(x[['Weight']])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
lr_predictions = pd.DataFrame({'y_pred': y_pred, 'y_test': y_test})
```

#  Prediction using Linear regression:-

```python
lr_predictions.sample(20)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```
