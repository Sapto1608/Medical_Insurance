# Drawing Histogram for Ages

```python
sns.histplot(data['Age'],bins=10,edgecolor='red',kde=True,stat='density')
```

```python
# Checking for skewness
print(data['Age'].skew())
print('Very slightly positively skewed')
```

# Drawing Histogram for Heights

```python
sns.histplot(data['Height'],bins=10,edgecolor='red',kde=True,stat='density')
```

```python
# Checking for skewness
print(data['Height'].skew())
print('Slightly negatively skewed')
```

# Drawing Histogram for Weights

```python
sns.histplot(data['Weight'],bins=10,edgecolor='red',kde=True,stat='density')
```

# Heatmap

```python
plt.figure(figsize=(14,8))
sns.heatmap(data.corr(), annot = True, cmap='coolwarm',linewidths=.1)
plt.title("Heatmap for correlation between columns")
plt.show()
```

# Pairplots

```python
sns.pairplot(data,hue = 'PremiumPrice',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
